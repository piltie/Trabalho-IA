Importações padrão:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
```

Gerando dataset:

```python

from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=3000, centers=4,
                       cluster_std=0.90, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

Treinando com 5 clusters e visualizando:

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

Treinando com 8 clusters e visualizando:

```python
kmeans2 = KMeans(n_clusters=8)
kmeans2.fit(X)
y_kmeans2 = kmeans2.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans2, s=50, cmap='viridis')

centers2 = kmeans2.cluster_centers_
plt.scatter(centers2[:, 0], centers2[:, 1], c='black', s=200, alpha=0.5);
```
